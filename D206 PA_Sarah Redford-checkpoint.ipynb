```python
# Create new dataframe to hold all continuous variables to be used for PCA
pca_df = df[['Income','Outage_sec_perweek','Tenure','MonthlyCharge','Bandwidth_GB_Year']]
```

```python
# Verify that new dataframe was created properly
pca_df.info()
```

```python
# Create a new datafram containing the normalized data from the pca dataset
pca_df_normalized = (pca_df-pca_df.mean())/pca_df.std()
```

```python
# Verify that the new datafram was created properly
pca_df_normalized.info()
```

```python
# Create pca dataframe and fit that dataframe into the normalized dataframe that was created before
pca = PCA(n_components=pca_df.shape[1])
```

```python
pca.fit(pca_df_normalized)
```

```python
# Manually create the dataframe with the below column names for the principal components
final_pca = pd.DataFrame(pca.transform(pca_df_normalized),
                        columns=['PC1','PC2','PC3','PC4','PC5'])
```

```python
# Verify the dataframe was created correctly
final_pca.info()
```

```python
# Display PCA Loadings
loadings = pd.DataFrame(pca.components_.T,
                        columns=['PC1','PC2','PC3','PC4','PC5'],
                        index=pca_df.columns)
loadings
```

```python
# Create correlation matrix plot for loadings
matrix = sns.heatmap(loadings, annot=True, cmap='Spectral')
plt.show()
```

```python
# Create scree plot to help determine which PC we should keep based on the eigenvalues
# calculate covariance and vectors before defining eigenvalues
cov_matrix = np.dot(pca_df_normalized.T, pca_df_normalized) / pca_df.shape[0]
```

```python
eigenvalues = [np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)) for eigenvector in pca.components_]
```

```python
# plot the eigenvalues in a scree plot 
plt.plot(eigenvalues)
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalues')
plt.axhline(y=1, color="green")
plt.show()
```
